```python
from data.parameters import XI_2, task_1_threshold
from task_1.calculations import calculate_poisson_probabilities, calculate_poisson_median
from task_1.vizs import generate_and_save_poisson_graph
```

```python
# Graph parameters
lambda_=XI_2
probabilities = calculate_poisson_probabilities(lambda_,threshold=task_1_threshold)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the median
median = calculate_poisson_median(lambda_)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(np.arange(len(probabilities)), probabilities)

ax.set_xlabel("Number of Events")
ax.set_ylabel("Probability")
ax.set_title(f"Poisson Distribution ($\lambda$ = {lambda_})")

ax.axvline(x=XI_2, color='red', linestyle='--', label='Expected Value')
ax.axvline(x=median, color='green', linestyle='--', label='Median')
ax.legend()
# save it in pdf in the folder figures
plt.savefig('figures/poisson_distribution.pdf')
```
